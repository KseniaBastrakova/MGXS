```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
number_iterations = [1000, 2000, 4000, 6000, 8000]
```

```python
difference_k_effective_1_cs = np.array([2.612870250027125, 2.6129587755379324, 2.6128999425215755 , 2.612776752189073, 2.61295907687632]) - 2.612903 
```

```python
print(difference_k_effective_1_cs)
```

```python
difference_k_effective_1_cs = difference_k_effective_1_cs * 100000
```

```python
print(difference_k_effective_1_cs)
```

```python
std_1_cs = [1.521954068313456e-07, 5.286662293366592e-08, 1.904570384854544e-08, 1.0597085420099517e-08, 6.823539539470438e-09]
```

```python
plt.figure(figsize=(10, 10))
plt.plot(number_iterations, difference_k_effective_1_cs, marker='o')


plt.xlabel('number of iterations')
plt.ylabel('difference in k effective in pcm')
```

```python
plt.figure(figsize=(10, 10))
plt.plot(number_iterations, std_1_cs, marker='o')


plt.xlabel('number of iterations')
plt.ylabel('standart deviation')
```
